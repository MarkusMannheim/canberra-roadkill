```python
import pandas as pd
import math
data = pd.read_csv("Wildlife_incident_locations.csv")
data.columns = ["date", "species", "road", "lat", "lng"]
data[["lat", "lng", "road", "species"]].to_csv("stats.csv", index=False)
data["date"] = pd.to_datetime(data["date"])
data.sort_values("date", inplace=True)
data["month"] = data["date"].dt.strftime("%b %Y")
data
```

```python
point = [data.loc[0]["lng"], data.loc[0]["lat"]]
distances = []
for index, row in data.iterrows():    
    distance = math.sqrt(((point[0] - row["lng"]) ** 2) + ((point[1] - row["lat"]) ** 2))
    distances.append(distance)
distances.sort()
distances
```
